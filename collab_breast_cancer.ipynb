```python
import pandas as pd
import numpy as np
import struct
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
#importing the svm model lib
from sklearn.svm import SVC
# import metric libraries
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('data.csv')
```

```python
# show all columns
pd.set_option('display.max_columns', None)
# print first 5 rows
df.head()
```

```python
# remove id column and unnamed column
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
# apply info function to check data
df.info()
```

```python
df.describe()
```

```python
# rename target column
df = df.rename(columns={'diagnosis': 'target'})
# target column replace with 0 and 1
df['target'] = df['target'].replace({'B': 0, 'M': 1})
```

```python
# Define a color palette (you can customize this list of colors)
colors = ["#1f77b4", "#ff7f0e"]

# Create the plot with different colors for each bar
ax = sns.countplot(x='target', data=df, palette=colors)

# Add count on each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(20, 17))
matrix = np.triu(df.corr())
sns.heatmap(df.corr(), annot=True, mask=matrix, linewidth=.8, cmap="rocket");
```

```python
X = np.random.randn(100, 2)
y = np.random.randint(0, 2, 100)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Plot the training and testing sets
plt.scatter(X_train[:, 0], X_train[:, 1], c='blue', label='Training set')
plt.scatter(X_test[:, 0], X_test[:, 1], c='red', label='Testing set')
plt.legend()
plt.show()
```

```python
print('the size of our "Y" is', y.shape)
print('\n')
print('the size of our training "Y" (input features) is', y_train.shape)
print('\n')
print('the size of our testing "Y" (input features) is',y_test.shape)
```

```python
print('the size of our "X" is', X.shape)
print('\n')
print('the size of our training "X" (input features) is', X_train.shape)
print('\n')
print('the size of our testing "X" (input features) is',X_test.shape)
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
y_predict = svc_model.predict(X_test)
```

```python
cm = np.array(confusion_matrix(y_test,y_predict, labels = [1,0]))
confusion = pd.DataFrame(cm, index = ['is_cancer', 'is_healthy'],
                        columns = ['predicted_cancer', 'predicted_healthy'])
confusion
```

```python
# Visualization on HeatMap
sns.heatmap(confusion, annot = True)
```

```python
print(classification_report(y_test, y_predict))
```
